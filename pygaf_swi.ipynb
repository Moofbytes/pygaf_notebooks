## Ghyben-Hertzberg Abrupt Interface Solution
The default `GhybenHertzberg` object uses the `Aq1dFiniteUnconf` aquifer class and the `Steady1dUnconfFlow` flow solution class with type 1 boundary condition at x=0 and type 2 boundary conditions at x=L.

Create a default Ghyben-Hertzberg interface solution and display the default solution information.

```python
SW1 = pygaf.GhybenHertzberg()
SW1.info()
```

Flow solution information.

```python
SW1.flow.info()
SW1.flow.aq.info()
SW1.flow.aq.draw()
```

Head solution.

```python
dum = SW1.flow.h()
```

Interface solution.

```python
dum = SW1.swi()
```

Change aquifer hydraulic conductivity.

```python
SW1.flow.aq.K = 2
dum = SW1.swi()
```

Change aquifer through flow, aquifer length and aquifer bottom elevation.

```python
SW1.flow.bcL.flow = -0.02
SW1.flow.aq.L = 300
SW1.flow.aq.bot = -120
SW1.flow.aq.info()
dum = SW1.flow.q()
dum = SW1.swi()
```
