## `MineSteadyRadLeaky` Class
Instanciate a defualt solution class and display information using the `.info` and `.aq.info` methods.

```python
S3.aq.K = 1.54
S3.Kleak = 1e-5
S3.aq.B = 10
S3.Bleak = 5
S3.h0 = 154
S3.rp = 127
S3.qp = 3000
```

```python
print('The aquitard leakage factor is:', round(S3.lfac, 1))
```

The radius of influence is the radius where drawdown is 0.1% of initial groundwater head.

```python
print('The radius of influence is:', round(S3.ri, 1))
```

### Radial drawdown method

```python
dd = S3.dd(n=20)
print(dd)
```

## `MineSteadyRadUnconf` Class
Instanciate a defualt solution class and display information using the `.info` and `.aq.info` methods.

```python
S1 = pygaf.MineSteadyRadUnconf()
S1.info()
S1.aq.info()
```

Display the definition diagram.

```python
S1.draw(dw=9)
```

Specifiy parameter values.

```python
S1.aq.K = 0.1
S1.aq.B = 154
S1.hp = 50
S1.rp = 127
S1.R = 50/1000/365
```

### Radius of influence property
The radius of influence is the radius of the recharge area contributing to steady mine inflow.

```python
print('The radius of influence is:', round(S1.ri, 1))
```

### Mine inflow property
Steady inflow rate to the mine pit.

```python
print('The steady pit inflow rate is:', round(S1.qp, 1))
```

```python
from numpy import linspace
import pandas
hp = [i for i in linspace(S1.aq.B-5, S1.aq.B/3, 5)]
dp = [S1.dp for S1.hp in hp]
qp = [S1.qp for S1.hp in hp]
df = pandas.DataFrame({'pit head':hp, 'pit drawdown':dp, 'pit inflow':qp})
print(df)
```

### Radial drawdown method

```python
dd = S1.dd(n=20)
print(dd.tail())
```

## `MineSteadyRadUnconf2` Class
Instanciate a default solution class and display information using the `.info` and `.aq.info` methods.

```python
S2 = pygaf.MineSteadyRadUnconf2()
S2.info()
S2.aq.info()
```

```python
S2.draw(dw=9)
```

Specifiy parameter values.

```python
S2.aq.K = 0.1
S2.aq.B = 154
S2.hp = 50
S2.rp = 127
S2.R = 50/1000/365
S2.aq2kx = 0.01
S2.aq2kz = 0.01
S2.D = 1
```

### Radius of influence property
The radius of influence is the radius of the recharge area contributing to steady mine inflow.

```python
print('The radius of influence is:', round(S2.ri, 1))
```

### Mine inflow property
Steady inflow rate to the mine pit.

```python
print('The steady pit inflow rate from the upper aquifer is:', round(S2.qp1, 1))
print('The steady pit inflow rate from the lower aquifer is:', round(S2.qp2, 1))
print('The steady total pit inflow rate is:', round(S2.qp, 1))
```

```python
from numpy import linspace
import pandas
hp = [i for i in linspace(S2.aq.B-5, S2.aq.B/3, 5)]
dp = [S2.dp for S2.hp in hp]
qp1 = [S2.qp1 for S2.hp in hp]
qp2 = [S2.qp2 for S2.hp in hp]
qp = [S2.qp for S2.hp in hp]
df = pandas.DataFrame({'pit head':hp, 'pit drawdown':dp, 'upper inflow':qp1, 'lower inflow':qp2, 'total inflow':qp})
print(df)
```

### Radial drawdown method

```python
dd = S2.dd(n=25)
print(dd.tail())
```
