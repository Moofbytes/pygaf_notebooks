The default `Steady1dConfFlow` object uses the default `Aq1dFiniteConf` aquifer class, default `SteadyBC` type 2 boundary condition class at x=0 and default `SteadyBC` type 1 boundary condition class at x=L.

```python
S = pygaf.Steady1dConfFlow()
S.info()
S.bcL.info()
S.aq.info()
S.aq.draw()
H = S.h()
Q = S.q()
G =S.h_grad()
```

### Head and no-flow BCs

```python
S1 = pygaf.Steady1dConfFlow()
S1.R = 0.0001
S1.info()
S1.h()
Q1 = S1.q()
G1 = S1.h_grad()
```

Swap the boundary conditions.

```python
S2 = pygaf.Steady1dConfFlow()
S2.bc0.type = 1
S2.bcL.type = 2
S2.R = 0.0001
S2.info()
H2 = S2.h()
Q2 = S2.q()
G2 = S2.h_grad()
```
